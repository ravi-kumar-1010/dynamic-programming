```python
## RECURSIVE SOLUTION
def decode(s):
    res = 0
    print(s)
    def dfs(i):
        if i == len(s):
            return 1
        if s[i] == '0':
            return 0
        res = dfs(i+1)
        if (i+1 < len(s) and (s[i] == '1' or s[i] == '2' and s[i+1] in '0123456')):
            res += dfs(i+2)
        return res  
    return dfs(0)
decode("111111111111111111111111111111111111111111111")
```

```python
## DYNAMIC SOLUTION
def decode(s):
    dp = {(len(s)):1}
    def dfs(i):
        # print(i,dp)
        if i in dp:
            return dp[i]    
        if s[i] == '0':
            return 0
        res = dfs(i+1)
        if (i+1 < len(s) and (s[i] == '1' or s[i] == '2' and s[i+1] in '0123456')):
            res += dfs(i+2)
        dp[i] = res
        # print("End: ",i,dp)
        return res
    return dfs(0)
decode("111111111111111111111111111111111111111111111")
```
