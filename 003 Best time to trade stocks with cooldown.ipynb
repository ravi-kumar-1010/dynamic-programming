```python
def maxProfit(arr):
    n = len(arr)
    profit = 0
    prev = 'cool'
    i = 0
    preval = 0 # previous value
    def stock(arr,i,prev,preval,n,profit):
        print("########: ",prev)
        if i >= n:  # if all elements are covered
            return profit
        elif prev == 'sell':   # if you sold previously you need to cooldown
            print('-----')
            print("sell")
            x = stock(arr,i+1,'cool',preval,n,profit) # notice you I didn't change previous value
            print(x)
            print('-----')
            return x   
        elif prev == 'buy':    # if you bought previously you can either sell or have a cooldown
            print("-----")
            print('Buy')
            x = stock(arr,i+1,'sell',0,n,profit+arr[i])
            y = stock(arr,i+1,'cool',preval,n,profit)
            print(x,y)
            print('-----')
            return max(x,y)
        elif prev == 'cool':   # if you are on a cooldown you can either buy or you can have a cooldown
            print('-----')
            print("Cool")
            x = stock(arr,i+1,'buy',arr[i],n,profit-arr[i])
            y = stock(arr,i+1,'cool',preval,n,profit)
            print(x,y)
            print('-----')
            return max(x,y)
            
    res = stock(arr,i,prev,preval,n,profit)
    return res
arr = [1,2,3,0,2]
print(maxProfit(arr))
```

```python
# DP Solution
def maxProfit(prices):
    dp = {}
    def dfs(i,buying):
        if i >= len(prices):
            return 0
        if (i,buying) in dp:
            return dp[(i,buying)]
        cooldown = dfs(i+1,buying)
        if buying:
            buy = dfs(i+1,not buying)-prices[i]
            dp[(i,buying)] = max(buy,cooldown)
        else:
            sell = dfs(i+2,not buying) +prices[i]
            dp[(i,buying)] = max(sell,cooldown)
        return dp[(i,buying)]
    return dfs(0,True)

print("Final Output",maxProfit([1,2,3,0,2]))
```
